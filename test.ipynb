# kakao map api 자동 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import requests; from urllib.parse import urlparse
def address_to_latitude(address):
    global lat
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+address
    result = requests.get(urlparse(url).geturl(),
                     headers = {"Authorization": "KakaoAK 78a23dac4d5a733b24bed6ab79cb0c4f"})
    json_obj = result.json()
    for document in json_obj['documents']:
        lat = document['y']
    return lat

def address_to_longtitude(address):
    global long
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+address
    result = requests.get(urlparse(url).geturl(),
                     headers = {"Authorization": "KakaoAK 78a23dac4d5a733b24bed6ab79cb0c4f"})
    json_obj = result.json()
    for document in json_obj['documents']:
         long = document['x']
    return long
# 마포 용산 강남 중구 종로구 강북구 도봉구

# petLoc = ["마포구", "용산구", "강남구", "중구", "종로구", "강북구", "도봉구"]
petLoc = ["마포구", "용산구"]
find = "동물병원"
results = []
time.sleep(10)

driver = webdriver.Chrome(
    executable_path= "/Users/ham/Code/Python/chromedriver"
)
url = "https://m.map.naver.com/"
driver.get(url)
action = ActionChains(driver)



for loc in petLoc:
    time.sleep(4)
    driver.find_element_by_class_name('Nbox_input_text').click()
    searchPetHos = loc + find
    driver.find_element_by_class_name('Nbox_input_text._search_input').send_keys(searchPetHos)
    driver.find_element_by_xpath('//*[@id="ct"]/div[1]/div[1]/form/div/div[2]/div/span[2]/button[2]').click()

    time.sleep(4)
    replys =driver.find_elements_by_xpath('//*[@id="ct"]/div[2]/ul/li')
    print(len(replys))
    #ct > div.search_listview._content._ctList > ul > li:nth-child(1) > div.item_info > div.item_info_inn > div > a
    #ct > div.search_listview._content._ctList > ul > li:nth-child(1) > div.item_info > div.item_info_inn > div > a
    #ct > div.search_listview._content._ctList > ul > li:nth-child(1) > div.item_info > a.a_item.a_item_distance._linkSiteview
    for index, reply in enumerate(replys):
            code = reply.find_element_by_css_selector('div.wrap_item > a.data-cid')
            name = reply.find_element_by_css_selector('div.item_tit').text
            address =reply.find_element_by_css_selector('div.wrap_item').text.split('\n')[1]
            latitude = address_to_latitude(address)
            longtitude = address_to_longtitude(address)
            results.append((name, address, latitude, longtitude))
            
    driver.back()

driver.close()
results
```
